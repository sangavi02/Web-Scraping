### Import the required Libraries

```python
import requests
import csv
from bs4 import BeautifulSoup
url = ['https://internshala.com/jobs/work-from-home/','https://internshala.com/jobs/jobs-in-delhi/','https://internshala.com/jobs/jobs-in-mumbai/','https://internshala.com/jobs/jobs-in-bangalore/','https://internshala.com/jobs/jobs-in-hyderabad/','https://internshala.com/jobs/jobs-in-kolkata/','https://internshala.com/jobs/jobs-in-chennai/','https://internshala.com/jobs/jobs-in-pune/','https://internshala.com/jobs/jobs-in-jaipur/']
```

### Web Scrapping Method

```python
def trail(url):
    
    # Send a GET request to the website
    try:
        response = requests.get(url)
        html_content = response.content
    except requests.exceptions.RequestException as e:
        # If there is an error with the request, the website is not available
        print(f"Website {url} is not available.")
        print(e)
    else:
        # If the request is successful, the website is available
        print(f"Website {url} is available.")
        # Extract data from the website here

    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all job postings
    job_postings = soup.find_all('div', {'class': 'individual_internship'})


    rows = []
    # Extract job titles, companies, and locations from each job posting
    for job in job_postings:
    
        title = job.find_all('h3', {'class': 'heading_4_5 profile'})
        if title is not None:
            title = title[0].text.strip()
        else:
            title = "Not available"
        
        company = job.find('h4', {'class': 'heading_6 company_name'})
        if company is not None:
            company = company.text.strip()
        else:
            company = "Not available"
        
        location = job.find('p', {'id': 'location_names'}).span
        if location is not None:
            location = location.text.strip()
        else:
            location = "Not available"
    
        salary_div = job.find('div', {'class': 'item_body salary'})
        if salary_div is not None and salary_div.find('i') is not None:
            salary = salary_div.find('i').next_sibling.strip()
        else:
            salary = "Not available"
        
        start_date_div = job.find('div', {'id': 'start-date-first'})
        if start_date_div is not None:
            start_date = start_date_div.text.strip()
        else:
            start_date = "Not available"
    
        # extract the actively hiring badge (if present)
        actively_hiring_badge = job.find('div', {'class': 'actively_hiring_badge'})
        if actively_hiring_badge is not None:
            status = actively_hiring_badge.text.strip()
        else:
            status = "Not actively hiring"
        
        
        print(f"Title: {title}")
        print(f"Company: {company}")
        print(f"Location: {location}")
        print(f"Salary: {salary}")
        print(f"Status: {status}")
        print(f"Start Date: {start_date}")
        print('---------------------------------------')
    
        rows.append([title, company, location , start_date, salary , status])
    
    # Save the extracted data to a CSV file
    with open('internship_data.csv', mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Title', 'Company', 'Location', 'Start Date', 'Salary', 'Status'])
        writer.writerows(rows)
    
    print("Data saved to internship_data.csv")
```

### Select the URL's to be scrapped

```python
print(url)
```

### Scrap and Store the data's from  Websites

```python
for url in url:
    trail(url)
```
